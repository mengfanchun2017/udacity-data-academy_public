### ç»ƒä¹ : å¼‚å¸¸å€¼æ£€æµ‹
å¯¹äºä»»ä½•çš„åˆ†æï¼Œåœ¨æ•°æ®é¢„å¤„ç†çš„è¿‡ç¨‹ä¸­æ£€æµ‹æ•°æ®ä¸­çš„å¼‚å¸¸å€¼éƒ½æ˜¯éå¸¸é‡è¦çš„ä¸€æ­¥ã€‚å¼‚å¸¸å€¼çš„å‡ºç°ä¼šä½¿å¾—æŠŠè¿™äº›å€¼è€ƒè™‘è¿›å»åç»“æœå‡ºç°å€¾æ–œã€‚è¿™é‡Œæœ‰å¾ˆå¤šå…³äºæ€æ ·å®šä¹‰ä»€ä¹ˆæ˜¯æ•°æ®é›†ä¸­çš„å¼‚å¸¸å€¼çš„ç»éªŒæ³•åˆ™ã€‚è¿™é‡Œæˆ‘ä»¬å°†ä½¿ç”¨[ Tukey çš„å®šä¹‰å¼‚å¸¸å€¼çš„æ–¹æ³•](http://datapigtechnologies.com/blog/index.php/highlighting-outliers-in-your-data-with-the-tukey-method/)ï¼šä¸€ä¸ªå¼‚å¸¸é˜¶ï¼ˆoutlier stepï¼‰è¢«å®šä¹‰æˆ1.5å€çš„å››åˆ†ä½è·ï¼ˆinterquartile rangeï¼ŒIQRï¼‰ã€‚ä¸€ä¸ªæ•°æ®ç‚¹å¦‚æœæŸä¸ªç‰¹å¾åŒ…å«åœ¨è¯¥ç‰¹å¾çš„ IQR ä¹‹å¤–çš„ç‰¹å¾ï¼Œé‚£ä¹ˆè¯¥æ•°æ®ç‚¹è¢«è®¤å®šä¸ºå¼‚å¸¸ç‚¹ã€‚

åœ¨ä¸‹é¢çš„ä»£ç å•å…ƒä¸­ï¼Œä½ éœ€è¦å®Œæˆä¸‹é¢çš„åŠŸèƒ½ï¼š
 - å°†æŒ‡å®šç‰¹å¾çš„ 25th åˆ†ä½ç‚¹çš„å€¼åˆ†é…ç»™ `Q1` ã€‚ä½¿ç”¨ `np.percentile` æ¥å®Œæˆè¿™ä¸ªåŠŸèƒ½ã€‚
 - å°†æŒ‡å®šç‰¹å¾çš„ 75th åˆ†ä½ç‚¹çš„å€¼åˆ†é…ç»™ `Q3` ã€‚åŒæ ·çš„ï¼Œä½¿ç”¨ `np.percentile` æ¥å®Œæˆè¿™ä¸ªåŠŸèƒ½ã€‚
 - å°†æŒ‡å®šç‰¹å¾çš„å¼‚å¸¸é˜¶çš„è®¡ç®—ç»“æœèµ‹å€¼ç»™ `step`ã€‚
 - é€‰æ‹©æ€§åœ°é€šè¿‡å°†ç´¢å¼•æ·»åŠ åˆ° `outliers` åˆ—è¡¨ä¸­ï¼Œä»¥ç§»é™¤å¼‚å¸¸å€¼ã€‚

**æ³¨æ„ï¼š** å¦‚æœä½ é€‰æ‹©ç§»é™¤å¼‚å¸¸å€¼ï¼Œè¯·ä¿è¯ä½ é€‰æ‹©çš„æ ·æœ¬ç‚¹ä¸åœ¨è¿™äº›ç§»é™¤çš„ç‚¹å½“ä¸­ï¼
ä¸€æ—¦ä½ å®Œæˆäº†è¿™äº›åŠŸèƒ½ï¼Œæ•°æ®é›†å°†å­˜å‚¨åœ¨ `good_data` ä¸­ã€‚

````python
# outlier_index
outliers = []

# å¯¹äºæ¯ä¸€ä¸ªç‰¹å¾ï¼Œæ‰¾åˆ°å€¼å¼‚å¸¸é«˜æˆ–è€…æ˜¯å¼‚å¸¸ä½çš„æ•°æ®ç‚¹
for feature in log_data.keys():
    
    # TODO: è®¡ç®—ç»™å®šç‰¹å¾çš„Q1ï¼ˆæ•°æ®çš„25thåˆ†ä½ç‚¹ï¼‰
    Q1 = np.percentile(log_data[feature], 25)
    
    # TODO: è®¡ç®—ç»™å®šç‰¹å¾çš„Q3ï¼ˆæ•°æ®çš„75thåˆ†ä½ç‚¹ï¼‰
    Q3 = np.percentile(log_data[feature], 75)
    
    # TODO: ä½¿ç”¨å››åˆ†ä½èŒƒå›´è®¡ç®—å¼‚å¸¸é˜¶ï¼ˆ1.5å€çš„å››åˆ†ä½è·ï¼‰
    step = 1.5*(Q3-Q1)
    
    # æ˜¾ç¤ºå¼‚å¸¸ç‚¹
    print("Data points considered outliers for the feature '{}':".format(feature))
    feature_outlier = log_data[~((log_data[feature] >= Q1 - step) & (log_data[feature] <= Q3 + step))]
    display(feature_outlier)
    print(feature_outlier.index.tolist())
    outliers = outliers + feature_outlier.index.tolist()
    
# TODO(å¯é€‰): é€‰æ‹©ä½ å¸Œæœ›ç§»é™¤çš„æ•°æ®ç‚¹çš„ç´¢å¼•
## outliers  = []
## æåˆ°å¾ªç¯å‰å¤„ç†

# IQR è®¡ç®—çš„è¯´æ˜
## https://machinelearningmastery.com/how-to-use-statistics-to-identify-outliers-in-data/
## https://www.khanacademy.org/math/statistics-probability/summarizing-quantitative-data/box-whisker-plots/a/identifying-outliers-iqr-rule
````

````python
# updated v2 find who to delete
seen = {}
dupes = []

for x in outliers:
    if x not in seen:
        seen[x] = 1
    else:
        if seen[x] == 1:
            dupes.append(x)
        seen[x] += 1

dupes

# nice eg
## https://stackoverflow.com/questions/9835762/how-do-i-find-the-duplicates-in-a-list-and-create-another-list-with-them
## do not use count(too slow)
````

````python
# ä»¥ä¸‹ä»£ç ä¼šç§»é™¤outliersä¸­ç´¢å¼•çš„æ•°æ®ç‚¹, å¹¶å‚¨å­˜åœ¨good_dataä¸­
good_data = log_data.drop(log_data.index[dupes]).reset_index(drop = True)
````

## ç‰¹å¾è½¬æ¢
åœ¨è¿™ä¸ªéƒ¨åˆ†ä¸­ä½ å°†ä½¿ç”¨ä¸»æˆåˆ†åˆ†æï¼ˆPCAï¼‰æ¥åˆ†ææ‰¹å‘å•†å®¢æˆ·æ•°æ®çš„å†…åœ¨ç»“æ„ã€‚ç”±äºä½¿ç”¨PCAåœ¨ä¸€ä¸ªæ•°æ®é›†ä¸Šä¼šè®¡ç®—å‡ºæœ€å¤§åŒ–æ–¹å·®çš„ç»´åº¦ï¼Œæˆ‘ä»¬å°†æ‰¾å‡ºå“ªä¸€ä¸ªç‰¹å¾ç»„åˆèƒ½å¤Ÿæœ€å¥½çš„æç»˜å®¢æˆ·ã€‚

### ç»ƒä¹ : ä¸»æˆåˆ†åˆ†æï¼ˆPCAï¼‰

æ—¢ç„¶æ•°æ®è¢«ç¼©æ”¾åˆ°ä¸€ä¸ªæ›´åŠ æ­£æ€åˆ†å¸ƒçš„èŒƒå›´ä¸­å¹¶ä¸”æˆ‘ä»¬ä¹Ÿç§»é™¤äº†éœ€è¦ç§»é™¤çš„å¼‚å¸¸ç‚¹ï¼Œæˆ‘ä»¬ç°åœ¨å°±èƒ½å¤Ÿåœ¨ `good_data` ä¸Šä½¿ç”¨PCAç®—æ³•ä»¥å‘ç°æ•°æ®çš„å“ªä¸€ä¸ªç»´åº¦èƒ½å¤Ÿæœ€å¤§åŒ–ç‰¹å¾çš„æ–¹å·®ã€‚é™¤äº†æ‰¾åˆ°è¿™äº›ç»´åº¦ï¼ŒPCA ä¹Ÿå°†æŠ¥å‘Šæ¯ä¸€ä¸ªç»´åº¦çš„è§£é‡Šæ–¹å·®æ¯”ï¼ˆexplained variance ratioï¼‰--è¿™ä¸ªæ•°æ®æœ‰å¤šå°‘æ–¹å·®èƒ½å¤Ÿç”¨è¿™ä¸ªå•ç‹¬çš„ç»´åº¦æ¥è§£é‡Šã€‚æ³¨æ„ PCA çš„ä¸€ä¸ªç»„æˆéƒ¨åˆ†ï¼ˆç»´åº¦ï¼‰èƒ½å¤Ÿè¢«çœ‹åšè¿™ä¸ªç©ºé—´ä¸­çš„ä¸€ä¸ªæ–°çš„â€œç‰¹å¾â€ï¼Œä½†æ˜¯å®ƒæ˜¯åŸæ¥æ•°æ®ä¸­çš„ç‰¹å¾æ„æˆçš„ã€‚

åœ¨ä¸‹é¢çš„ä»£ç å•å…ƒä¸­ï¼Œä½ å°†è¦å®ç°ä¸‹é¢çš„åŠŸèƒ½ï¼š
 - å¯¼å…¥ `sklearn.decomposition.PCA` å¹¶ä¸”å°† `good_data` ç”¨ PCA å¹¶ä¸”ä½¿ç”¨6ä¸ªç»´åº¦è¿›è¡Œæ‹Ÿåˆåçš„ç»“æœä¿å­˜åˆ° `pca` ä¸­ã€‚
 - ä½¿ç”¨ `pca.transform` å°† `log_samples` è¿›è¡Œè½¬æ¢ï¼Œå¹¶å°†ç»“æœå­˜å‚¨åˆ° `pca_samples` ä¸­ã€‚

````python
from sklearn.decomposition import PCA

# TODOï¼šé€šè¿‡åœ¨good dataä¸Šè¿›è¡ŒPCAï¼Œå°†å…¶è½¬æ¢æˆ6ä¸ªç»´åº¦
pca = PCA(n_components=6)
pca.fit(good_data)

# TODOï¼šä½¿ç”¨ä¸Šé¢çš„PCAæ‹Ÿåˆå°†å˜æ¢æ–½åŠ åœ¨log_samplesä¸Š
pca_samples = pca.transform(log_samples)
## æ­¤å¤„ä¼šæŠ¥é”™ï¼Œå› ä¸ºlog_samplesåœ¨å‰é¢å»æ‰äº†ä¸€ä¸ªfeature
## æ‰€ä»¥goodå’Œsampleçš„shapeä¸ä¸€æ ·
## å…ˆæ³¨é‡Šæ‰ï¼Œåç»­ä½¿ç”¨æ—¶å†è°ƒæ•´

# ç”ŸæˆPCAçš„ç»“æœå›¾
pca_results = vs.pca_results(good_data, pca)
````

### ç»ƒä¹ ï¼šé™ç»´
å½“ä½¿ç”¨ä¸»æˆåˆ†åˆ†æçš„æ—¶å€™ï¼Œä¸€ä¸ªä¸»è¦çš„ç›®çš„æ˜¯å‡å°‘æ•°æ®çš„ç»´åº¦ï¼Œè¿™å®é™…ä¸Šé™ä½äº†é—®é¢˜çš„å¤æ‚åº¦ã€‚å½“ç„¶é™ç»´ä¹Ÿæ˜¯éœ€è¦ä¸€å®šä»£ä»·çš„ï¼šæ›´å°‘çš„ç»´åº¦èƒ½å¤Ÿè¡¨ç¤ºçš„æ•°æ®ä¸­çš„æ€»æ–¹å·®æ›´å°‘ã€‚å› ä¸ºè¿™ä¸ªï¼Œ**ç´¯è®¡è§£é‡Šæ–¹å·®æ¯”ï¼ˆcumulative explained variance ratioï¼‰**å¯¹äºæˆ‘ä»¬ç¡®å®šè¿™ä¸ªé—®é¢˜éœ€è¦å¤šå°‘ç»´åº¦éå¸¸é‡è¦ã€‚å¦å¤–ï¼Œå¦‚æœå¤§éƒ¨åˆ†çš„æ–¹å·®éƒ½èƒ½å¤Ÿé€šè¿‡ä¸¤ä¸ªæˆ–è€…æ˜¯ä¸‰ä¸ªç»´åº¦è¿›è¡Œè¡¨ç¤ºçš„è¯ï¼Œé™ç»´ä¹‹åçš„æ•°æ®èƒ½å¤Ÿè¢«å¯è§†åŒ–ã€‚

åœ¨ä¸‹é¢çš„ä»£ç å•å…ƒä¸­ï¼Œä½ å°†å®ç°ä¸‹é¢çš„åŠŸèƒ½ï¼š
 - å°† `good_data` ç”¨ä¸¤ä¸ªç»´åº¦çš„PCAè¿›è¡Œæ‹Ÿåˆï¼Œå¹¶å°†ç»“æœå­˜å‚¨åˆ° `pca` ä¸­å»ã€‚
 - ä½¿ç”¨ `pca.transform` å°† `good_data` è¿›è¡Œè½¬æ¢ï¼Œå¹¶å°†ç»“æœå­˜å‚¨åœ¨ `reduced_data` ä¸­ã€‚
 - ä½¿ç”¨ `pca.transform` å°† `log_samples` è¿›è¡Œè½¬æ¢ï¼Œå¹¶å°†ç»“æœå­˜å‚¨åœ¨ `pca_samples` ä¸­ã€‚

````python
# TODOï¼šé€šè¿‡åœ¨good dataä¸Šè¿›è¡ŒPCAï¼Œå°†å…¶è½¬æ¢æˆä¸¤ä¸ªç»´åº¦
pca = PCA(n_components=2)
pca.fit(good_data)

# TODOï¼šä½¿ç”¨ä¸Šé¢è®­ç»ƒçš„PCAå°†good dataè¿›è¡Œè½¬æ¢
reduced_data = pca.transform(good_data)

# å…³äºè½¬æ¢
# https://www.researchgate.net/post/How_important_is_transforming_the_data_before_PCA

# TODOï¼šä½¿ç”¨ä¸Šé¢è®­ç»ƒçš„PCAå°†log_samplesè¿›è¡Œè½¬æ¢
pca_samples = pca.transform(log_samples)

# ä¸ºé™ç»´åçš„æ•°æ®åˆ›å»ºä¸€ä¸ªDataFrame
reduced_data = pd.DataFrame(reduced_data, columns = ['Dimension 1', 'Dimension 2'])
````

é€šè¿‡èšç±»æŠ€æœ¯ï¼Œæˆ‘ä»¬èƒ½å¤Ÿå°†åŸæœ‰çš„æ²¡æœ‰æ ‡è®°çš„æ•°æ®é›†ä¸­çš„é™„åŠ ç»“æ„åˆ†æå‡ºæ¥ã€‚å› ä¸ºæ¯ä¸€ä¸ªå®¢æˆ·éƒ½æœ‰ä¸€ä¸ªæœ€ä½³çš„åˆ’åˆ†ï¼ˆå–å†³äºä½ é€‰æ‹©ä½¿ç”¨çš„èšç±»ç®—æ³•ï¼‰ï¼Œæˆ‘ä»¬å¯ä»¥æŠŠç”¨æˆ·åˆ†ç±»ä½œä¸ºæ•°æ®çš„ä¸€ä¸ª[å·¥ç¨‹ç‰¹å¾](https://en.wikipedia.org/wiki/Feature_learning#Unsupervised_feature_learning)ã€‚å‡è®¾æ‰¹å‘å•†æœ€è¿‘è¿æ¥åä½æ–°é¡¾å®¢ï¼Œå¹¶ä¸”ä»–å·²ç»ä¸ºæ¯ä½é¡¾å®¢æ¯ä¸ªäº§å“ç±»åˆ«å¹´åº¦é‡‡è´­é¢è¿›è¡Œäº†é¢„ä¼°ã€‚è¿›è¡Œäº†è¿™äº›ä¼°ç®—ä¹‹åï¼Œæ‰¹å‘å•†è¯¥å¦‚ä½•è¿ç”¨å®ƒçš„é¢„ä¼°å’Œéç›‘ç£å­¦ä¹ çš„ç»“æœæ¥å¯¹è¿™åä¸ªæ–°çš„å®¢æˆ·è¿›è¡Œæ›´å¥½çš„é¢„æµ‹ï¼Ÿ

**æç¤º**ï¼šåœ¨ä¸‹é¢çš„ä»£ç å•å…ƒä¸­ï¼Œæˆ‘ä»¬æä¾›äº†ä¸€ä¸ªå·²ç»åšå¥½èšç±»çš„æ•°æ®ï¼ˆèšç±»ç»“æœä¸ºæ•°æ®ä¸­çš„clusterå±æ€§ï¼‰ï¼Œæˆ‘ä»¬å°†åœ¨è¿™ä¸ªæ•°æ®é›†ä¸Šåšä¸€ä¸ªå°å®éªŒã€‚å°è¯•è¿è¡Œä¸‹é¢çš„ä»£ç çœ‹çœ‹æˆ‘ä»¬å°è¯•é¢„æµ‹â€˜Regionâ€™çš„æ—¶å€™ï¼Œå¦‚æœå­˜åœ¨èšç±»ç‰¹å¾'cluster'ä¸ä¸å­˜åœ¨ç›¸æ¯”å¯¹æœ€ç»ˆçš„å¾—åˆ†ä¼šæœ‰ä»€ä¹ˆå½±å“ï¼Ÿè¿™å¯¹ä½ æœ‰ä»€ä¹ˆå¯å‘ï¼Ÿ

````python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split

# è¯»å–åŒ…å«èšç±»ç»“æœçš„æ•°æ®
cluster_data = pd.read_csv("cluster.csv")
y = cluster_data['Region']
X = cluster_data.drop(['Region'], axis = 1)

# åˆ’åˆ†è®­ç»ƒé›†æµ‹è¯•é›†
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=24)

clf = RandomForestClassifier(n_estimators=10, random_state=24)
clf.fit(X_train, y_train)
score_with_cluster = clf.score(X_test, y_test)

# ç§»é™¤clusterç‰¹å¾
X_train = X_train.copy()
X_train.drop(['cluster'], axis=1, inplace=True)
X_test = X_test.copy()
X_test.drop(['cluster'], axis=1, inplace=True)
clf.fit(X_train, y_train)
score_no_cluster = clf.score(X_test, y_test)

print("ä¸ä½¿ç”¨clusterç‰¹å¾çš„å¾—åˆ†: %.6f"%score_no_cluster)
print("ä½¿ç”¨clusterç‰¹å¾çš„å¾—åˆ†: %.6f"%score_with_cluster)

## è®¾ç½®äº†n_estimatoræ¶ˆé™¤äº†future warning
## å½“è®¾ä¸º100æ—¶ç»“æœä¸ºï¼š
## 0.724138 vs 0.724138
````
