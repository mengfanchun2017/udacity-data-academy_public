#### // check_info(df,list)

```python
# build function
def check_info(df,list):
    c = 1
    for i in list:
        print(('\n- columns #{} : {:-<8}'.format(c,i)))
        print((df[i].value_counts().nlargest(5)))
        ### nlargest ÈùûÂ∏∏Â•ΩÁî®
        c += 1
    pp.pprint('----checking complete----')
```
